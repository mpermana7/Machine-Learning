```python
from sklearn.cluster import MiniBatchKMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Membuat dataset acak untuk demonstrasi
X, _ = make_blobs(n_samples=1000, centers=4, cluster_std=0.6)

# Membuat model MiniBatchKMeans dengan 4 klaster
kmeans = MiniBatchKMeans(n_clusters=4, batch_size=100, random_state=42)
kmeans.fit(X)

# Prediksi klaster untuk setiap titik data
labels = kmeans.predict(X)

# Mendapatkan pusat klaster
centers = kmeans.cluster_centers_

# Plot data dan pusat klaster
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', alpha=0.5)
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x')
plt.title('MiniBatchKMeans Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
